# Linear Model

```python
# y=10+4x
lm=Models.LinearModel(
    n_params=2,
    params=[10,4],
    n_vars=1,
    n_obs=100,
    x_obs=[[x] for x in range(100)]
    )
print(lm)
```

```python
lm.plot_obs()
```

# Exponential Model

```python
#y=2e^(x/2)
em=Models.ExponentialModel(
    params=[2,.5],
    n_obs=20,
    x_obs=[[x] for x in np.linspace(0,5,20)]
    )
print(em)
```

```python
em.plot_obs()
```

# SIR Model

```python
population_size=100000
initially_infected=100
beta=1
gamma=.5

sm=Models.SIRModel(
    params=[population_size,initially_infected,beta,gamma],
    n_obs=30,
    x_obs=[[x] for x in range(30)]
    )
print(sm)
```

```python
sm.plot_obs()
```

# Stochastic SIR Model

```python
population_size=100000
initially_infected=100
beta=1
gamma=.5
alpha=.1

stochastic_sm=Models.StochasticSIRModel(
    params=(population_size,initially_infected,beta,gamma,alpha),
    n_obs=30,
    x_obs=[[x] for x in range(30)]
    )
print(stochastic_sm)
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

ax=Plotting.stochastic_sir_model_realisations(ax,stochastic_sm,n_reals=100)

plt.show()
```

# Gaussian Mixtures Model

```python
mu_1=-10; mu_2=20
sd_1=5; sd_2=1
pct=.3

gmm=Models.GaussianMixtureModel_two(
    params=(mu_1,mu_2,pct),
    n_obs=30,
    sd=(sd_1,sd_2)
)
print(gmm)
```
